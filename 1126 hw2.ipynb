###  1.裝飾器練習

```python
def test(func):
    def new_func(*args, **kwargs):
        print('start')
        result = func(*args, **kwargs)
        print('end')
        return result
    return new_func

@test
def greeting():
    print("Greeting, Earthing")
    
greeting()
```

### 2.選擇權畫圖練習

```python
x_4 = [11200,11400,11512,11700]
y_4 = [-112,-112,0,188]
plt.plot(x_4,y_4,'b--')
x_5 = [11200,11500,11554,11700]
y_5 = [108,108,0,-292]
plt.plot(x_5,y_5,'b--')
x_6 = [11200,11600,11619.5,11700]
y_6 = [-19.5,-19.5,0,80.5]
plt.plot(x_6,y_6,'b--')
result_x = [11200,11400,11500,11619.5,11700]
result_y = [-23.5,-23.5,76.5,-23.5,-23.5]
plt.plot(result_x,result_y,'r',linewidth=1.5)
plt.xlabel("strike price")
plt.ylabel("profit")
plt.title("break even chart")
```
