```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.datasets import load_boston
```

```python
boston_dataset = load_boston()
```

```python
boston = pd.DataFrame(boston_dataset.data, columns = boston_dataset.feature_names)
```

```python
boston['MEDV'] = boston_dataset.target
```

```python
import seaborn as sns
```

```python
X = boston
```

```python
X.drop("ZN",axis =1, inplace = True)
X.drop("CHAS", axis = 1, inplace = True)
X.drop("RAD", axis = 1, inplace = True)
X.drop("LSTAT", axis = 1, inplace = True)
X.drop("RM", axis = 1, inplace = True)
X.drop("INDUS", axis = 1, inplace = True)
```

```python
X.head()
```

```python
sns.distplot(X.MEDV, bins=50)
```

```python
Y = boston.MEDV
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,  y_train)
```

```python
y_predict = model.predict(x_test)
```

```python
sns.set()
plt.scatter(y_test, y_predict)
plt.xlim(0,60)
plt.ylim(0,60)
plt.plot([0,60],[0,60],'r')
#why my predict is in a line  QQ
```
